```python
import sys
sys.path.insert(0, '..')

import json
import re

import projects_parser
```

```python
BOOKLET_FILE = '../../booklets_data/2018.json'
```

```python
with open(BOOKLET_FILE, encoding='utf-8') as f:
    data = json.load(f)
```

```python
data
```

```python
pattern_full_stop = re.compile(r'(?<!Node)\.(?=\S)(?!NET)')
pattern_comma = re.compile(r'(?<!Node)\,(?=\S)(?!NET)')
```

```python
def normalize_value(value):
    stop_value = pattern_full_stop.subn('. ', value)[0]
    comma_value = pattern_comma.subn(', ', stop_value)[0]
    return comma_value
```

```python
FIELDS_TO_NORMALIZE = ['area', 'description', 'technology', 'area']
```

```python
normalized_data = []
for project in data:
    normalized_project = {**project}
    for field in FIELDS_TO_NORMALIZE:
        normalized_project[field] = normalize_value(normalized_project[field])
    normalized_data.append(normalized_project)
```

```python
projects_parser.write_json('2018', normalized_data, '../../booklets_data')
```
